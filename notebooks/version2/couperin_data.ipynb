```python
import pandas as pd
import html
import re

df = pd.read_parquet("MASTER_APC_ALL_20260211_003950.parquet")

# decode HTML entities (handles &#...; and &name;)
def unescape_html(x):
    if pd.isna(x) or not isinstance(x, str):
        return x
    # sometimes data is double-encoded; decode until stable (max 3 passes)
    for _ in range(3):
        y = html.unescape(x)
        if y == x:
            break
        x = y
    return x

# Option B: auto-detect any text column containing HTML entities and fix them
entity_pat = re.compile(r"&(#\d+|#x[0-9A-Fa-f]+|\w+);")
obj_cols = df.select_dtypes(include="object").columns
cols_to_fix = [c for c in obj_cols if df[c].dropna().astype(str).str.contains(entity_pat).any()]
for c in cols_to_fix:
    df[c] = df[c].map(unescape_html)

# Export (utf-8-sig helps Excel show accents correctly)
df.to_csv("all_couperin.csv", index=False, encoding="utf-8-sig")
```
